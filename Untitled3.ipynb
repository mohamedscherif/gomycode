```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
#this function is a function that gives us a plot of the correlation map by taking in it's process subplots, a diverging palette that allows the correlation map to have many colors in addition to a heatmap. So this function returns a plot correlation map after being runned.
```
